# REM:$\\$ $E(\sigma)$ ~ $\mathcal{N}(0,1)$ $\forall \sigma$

```python
n = 12
n_replicas = 1000
E = np.zeros((n_replicas, 2**n))
for i in range(n_replicas):
    E[i] = np.random.standard_normal(2**n)
    E[i] = sorted(E[i])
```

Energy density corresponds to the CDF of the standard normal distribution.

```python
plt.plot(np.arange(2**n), sorted(E[0]))
plt.show()
```

```python
plt.hist(np.mean(E, axis=0), bins=100)
plt.show()
```
