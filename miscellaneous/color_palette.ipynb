```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Define colors with names
colors = {
    "red": "#e41a1c",
    "blue": "#377eb8",
    "green": "#4daf4a",
    "purple": "#984ea3",
    "orange": "#ff7f00",
    "yellow": "#ffff33",
    "brown": "#a65628",
    "pink": "#f781bf",
    "grey": "#999999",
    "black": "#000000",
    "white": "#ffffff"
}

# Plot color palette
fig, ax = plt.subplots(figsize=(10, 2))
for i, (color_name, hex_value) in enumerate(colors.items()):
    ax.add_patch(mpatches.Rectangle((i, 0), 1, 1, color=hex_value))
    ax.text(i + 0.5, -0.1, color_name, ha='center', va='top', fontsize=10)

# Customizations
ax.set_xlim(0, len(colors))
ax.set_ylim(-0.3, 1)
ax.axis('off')

plt.show()
```
