```python
num_random_instances = 100
max_qubits = 50

instances = {}

for n in range(2,max_qubits+1):
    instances[n] = []
    for i in range(num_random_instances):
        arr = np.random.standard_normal(n + int(n*(n-1)/2)).tolist()
        instances[n].append(arr)
```

```python
# Save the dictionary as JSON
with open(projectdir+'instances.json', 'w') as f:
    json.dump(instances, f)
```
