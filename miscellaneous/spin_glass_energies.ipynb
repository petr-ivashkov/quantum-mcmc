# Spin glass: $\\$ $J_{ij}$ ~ $\mathcal{N}(0,1)$ $\forall i,j$ $\\$ E($\sigma$) = $\sum_{<ij>}J_{ij}\sigma_i\sigma_j$

```python
n = 12
n_replicas = 100
E = np.zeros((n_replicas, 2**n))
for i in range(n_replicas):
    m = RandomIsingModel(n=n, local_fields=False)
    E[i] = sorted(m.E)
```

Energy density corresponds (?) to the CDF of the normal distribution.

```python
plt.plot(np.arange(2**n), sorted(E[0]))
plt.show()
```

```python
plt.hist(np.mean(E, axis=0), bins=100)
plt.show()
```
