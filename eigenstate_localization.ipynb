```python
gamma_range = np.linspace(0,2,20)
p_vals = np.zeros_like(gamma_range)
for i, gamma in tqdm(enumerate(gamma_range)):
    H_ising = np.diag(m.E_rescaled)
    H_mixer = H_mixer_list[m.n-1]
    H = (1-gamma)*H_ising + gamma*H_mixer
    vals, vecs = la.eigh(H)
    # p = np.mean([vec**4 for vec in vecs.T])
    p = np.sum(vecs.T[0]**4)
    p_vals[i] = p
```

```python
plt.plot(gamma_range, p_vals)
```
