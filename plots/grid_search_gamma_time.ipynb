```python
grid_searches = {}
delta_quantum_arrays = {}

for n in range(5,11):
    grid_searches[n] = load_from_json(f'data/grid_search_n{n}.json')
    delta_quantum_arrays[n] = np.mean(grid_searches[n]['delta'], axis=2)

gamma_range = grid_searches[10]['gamma_range']
time_range = grid_searches[10]['time_range']
T = grid_searches[10]['T']
num_random_models = grid_searches[10]['num_random_models']
```

```python
from matplotlib.colors import LinearSegmentedColormap

# Define the colors for the custom colormap
colors = [
          (1.0, 1.0, 1.0),  # white
          (1.0, 1.0, 1.0), #'#f0d3d3',  # light grey
          red]  # red (default color of coolwarm colormap)

# Create a custom colormap
cmap_name = 'grey_red_coolwarm'
grey_red_coolwarm = LinearSegmentedColormap.from_list(cmap_name, colors)
```

```python
# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(figure_size_x, figure_size_y))

# Plot data on each subplot

pc00 = axs[0, 0].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[5], 
                        cmap=grey_red_coolwarm,
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[5])
                        # vmax=0.065,
                        )
pc01 = axs[0, 1].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[6], 
                        cmap=grey_red_coolwarm,
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[6])
                        # vmax=0.065,
                        )
pc02 = axs[0, 2].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[7], 
                        cmap=grey_red_coolwarm,
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[7])
                        # vmax=0.065,
                        )
pc10 = axs[1, 0].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[8], 
                        cmap=grey_red_coolwarm,
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[8])
                        # vmax=0.04,
                        )
pc11 = axs[1, 1].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[9], 
                        cmap=grey_red_coolwarm,
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[9])
                        # vmax=0.04,
                        )
pc12 = axs[1, 2].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[10], 
                        cmap=grey_red_coolwarm,
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[10]),
                        # vmax=0.04,
                        )
fig.colorbar(pc00)
fig.colorbar(pc01)
fig.colorbar(pc02)
fig.colorbar(pc10)
fig.colorbar(pc11)
fig.colorbar(pc12)

# Set labels
axs[1, 0].set_xlabel('$t$')
axs[1, 1].set_xlabel('$t$')
axs[1, 2].set_xlabel('$t$')

axs[0, 0].set_ylabel('$\gamma$')
axs[1, 0].set_ylabel('$\gamma$')

# Set ticks to []
axs[0, 0].set_xticks([])
axs[0, 1].set_xticks([])
axs[0, 2].set_xticks([])

axs[0, 1].set_yticks([])
axs[0, 2].set_yticks([])
axs[1, 1].set_yticks([])
axs[1, 2].set_yticks([])

axs[0, 0].set_title('n=5')
axs[0, 1].set_title('n=6')
axs[0, 2].set_title('n=7')
axs[1, 0].set_title('n=8')
axs[1, 1].set_title('n=9')
axs[1, 2].set_title('n=10')


# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()

fig.savefig(f'figures/grid_search_gamma_time_n5_10')
```

```python
import matplotlib.colors as mcolors

def lighten_color(color, amount=0.5):
    """
    Lightens the given color by interpolating it with white.

    Parameters:
    color (str): The color to lighten, in any Matplotlib-supported format.
    amount (float): The amount to lighten the color, from 0 (no change) to 1 (white).

    Returns:
    str: The lightened color in hexadecimal format.
    """
    try:
        c = mcolors.cnames[color]
    except KeyError:
        c = color
    c = mcolors.ColorConverter.to_rgb(c)
    white = np.array([1, 1, 1])
    lightened_color = (1 - amount) * np.array(c) + amount * white
    return mcolors.to_hex(lightened_color)

# Original color
red = '#B22222'

# Lightened color
lightened_red = lighten_color(red, amount=0.5)

print("Original color:", red)
print("Lightened color:", lightened_red)
```
