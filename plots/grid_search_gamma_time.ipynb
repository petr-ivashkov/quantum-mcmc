```python
grid_searches = {}
delta_quantum_arrays = {}

for n in range(4,10):
    grid_searches[n] = load_from_json(f'data/grid_search_n{n}.json')
    delta_quantum_arrays[n] = np.mean(grid_searches[n]['delta'], axis=2)

gamma_range = grid_searches[4]['gamma_range']
time_range = grid_searches[4]['time_range']
T = grid_searches[4]['T']
num_random_models = grid_searches[4]['num_random_models']
```

```python
# Create a 2x3 grid of subplots
fig, axs = plt.subplots(2, 3, figsize=(figure_size_x, figure_size_y))

# Plot data on each subplot

pc00 = axs[0, 0].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[4], 
                        cmap='coolwarm',
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[4])
                        # vmax=0.065,
                        )
pc01 = axs[0, 1].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[5], 
                        cmap='coolwarm',
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[5])
                        # vmax=0.065,
                        )
pc02 = axs[0, 2].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[6], 
                        cmap='coolwarm',
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[6])
                        # vmax=0.065,
                        )
pc10 = axs[1, 0].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[7], 
                        cmap='coolwarm',
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[7])
                        # vmax=0.04,
                        )
pc11 = axs[1, 1].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[8], 
                        cmap='coolwarm',
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[8])
                        # vmax=0.04,
                        )
pc12 = axs[1, 2].pcolor(time_range, 
                        gamma_range,
                        delta_quantum_arrays[9], 
                        cmap='coolwarm',
                        vmin=0,
                        vmax=np.max(delta_quantum_arrays[9]),
                        # vmax=0.04,
                        )
fig.colorbar(pc00)
fig.colorbar(pc01)
fig.colorbar(pc02)
fig.colorbar(pc10)
fig.colorbar(pc11)
fig.colorbar(pc12)

# Set labels
axs[1, 0].set_xlabel('$t$')
axs[1, 1].set_xlabel('$t$')
axs[1, 2].set_xlabel('$t$')

axs[0, 0].set_ylabel('$\gamma$')
axs[1, 0].set_ylabel('$\gamma$')

# Set ticks to []
axs[0, 0].set_xticks([])
axs[0, 1].set_xticks([])
axs[0, 2].set_xticks([])

axs[0, 1].set_yticks([])
axs[0, 2].set_yticks([])
axs[1, 1].set_yticks([])
axs[1, 2].set_yticks([])

axs[0, 0].set_title('n=4')
axs[0, 1].set_title('n=5')
axs[0, 2].set_title('n=6')
axs[1, 0].set_title('n=7')
axs[1, 1].set_title('n=8')
axs[1, 2].set_title('n=9')


# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()

fig.savefig(local_path+f'figures/grid_search_gamma_time_n4_9')
```
