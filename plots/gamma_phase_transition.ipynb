```python
n = 8
```

```python
grid_search = load_from_json(f'data/grid_search_n{n}.json')
```

```python
delta_quantum_arr = grid_search['delta']
gamma_range = grid_search['gamma_range']
time_range = grid_search['time_range']
T = grid_search['T']
num_random_models = grid_search['num_random_models']
```

```python
delta_quantum_arr_avg = np.mean(delta_quantum_arr, axis=2)
delta_quantum_arr_std = np.std(delta_quantum_arr, axis=2)
```

```python
fig = plt.figure(figsize=(figure_size_x, figure_size_y))

plt.plot(gamma_range, 
         np.mean(delta_quantum_arr_avg[:, 5:50], axis=1),
         '-',
         color=red,
         alpha=0.8)
plt.xlabel('$\gamma$')
plt.ylabel('$\delta$')
# plt.yticks([])

fig.tight_layout()

fig.savefig(f'figures/gamma_phase_transition_{n}')
```

```python
fig = plt.figure(figsize=(figure_size_x, figure_size_y))

pc0 = plt.pcolor(time_range, 
                gamma_range,
                delta_quantum_arr_avg, 
                cmap='coolwarm',
                vmin=0,
                vmax=np.max(delta_quantum_arr_avg)
                )
fig.colorbar(pc0)

plt.xlabel('$t$')
plt.ylabel('$\gamma$')
plt.title('$\delta$')

fig.tight_layout()

fig.savefig(f'figures/grid_search_n{n}')
```
